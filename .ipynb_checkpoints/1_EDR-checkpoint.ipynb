```python
# 1. Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: for better visuals
sns.set(style="whitegrid")
%matplotlib inline

# 2. Load the CSV file
df = pd.read_csv('dataset/heg_sample_data.csv')

# Show the shape and preview the dataset
print("Dataset shape:", df.shape)
```

```python
# Show first 10 rows
df.head(10)
```

```python
# 3. Check data types and missing values
df.info()
df.isnull().sum()
```

```python
# 4. Basic statistics (numerical columns)
df.describe()
```

```python
# Replace 'outcome' with your actual column name
df['outcome_label'] = df['outcome'].apply(lambda x: 'Death' if x == 1 else 'Live')

# Now plot the distribution
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='outcome_label', data=df, order=['Live', 'Death'])
plt.title('Distribution of Patient Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```
